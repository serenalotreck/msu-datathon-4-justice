```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
grid = pd.read_csv('../original_data/grids.csv')
grid = grid[ grid['grid'] == 'standard' ]
grid
```

```python
grid_dict = {}
for i in grid.index:
    key_ = str(grid.loc[i,'year']) + '_' + str(grid.loc[i,'severity']) + '_' + str(grid.loc[i,'chs'])
    grid_dict[key_] = int(grid.loc[i,'mid'])
```

```python
import json

with open('../generated_data/grid_dictionary.json', 'w') as fp:
    json.dump(grid_dict, fp)
```
